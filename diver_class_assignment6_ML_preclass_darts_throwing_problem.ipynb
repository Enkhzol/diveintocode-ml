**[Problem 3] Throw a lot of darts and record them in a state of being tied to the number**

```python
mean=[0,0]
cov=[[0.01,0],[0,0.01]] 

#target numbers of 1000 throws
targets1000 = {}
for number in range(1000):
  target_number = np.random.choice(6) + 1
  targets1000[str(number)] = target_number

print(targets1000)

#1000 throws in (X, Y) 
throws1000 = {}
for number in range(1000):
  xy = np.random.multivariate_normal(mean, cov)
  throws1000[str(number)] = xy

print(throws1000)
```

**[Problem 5] Functionalization**

```python
import matplotlib.pyplot as plt 

figure, axes = plt.subplots() 
circle10 = plt.Circle(( 0.0 , 0.0 ), 1.0, color = "blue" ) 
circle9 = plt.Circle(( 0.0 , 0.0 ), 0.9, color = "white" ) 
circle8 = plt.Circle(( 0.0 , 0.0 ), 0.8, color = "blue" ) 
circle7 = plt.Circle(( 0.0 , 0.0 ), 0.7, color = "white" ) 
circle6 = plt.Circle(( 0.0 , 0.0 ), 0.6, color = "blue" ) 
circle5 = plt.Circle(( 0.0 , 0.0 ), 0.5, color = "white" )
circle4 = plt.Circle(( 0.0 , 0.0 ), 0.4, color = "blue" ) 
circle3 = plt.Circle(( 0.0 , 0.0 ), 0.3, color = "white" )
circle2 = plt.Circle(( 0.0 , 0.0 ), 0.2, color = "blue" ) 
circle1 = plt.Circle(( 0.0 , 0.0 ), 0.1, color = "white" )
  
axes.set_aspect( 1 ) 

axes.add_artist( circle10 ) 
axes.add_artist( circle9 ) 
axes.add_artist( circle8 ) 
axes.add_artist( circle7 ) 
axes.add_artist( circle6 ) 
axes.add_artist( circle5 ) 
axes.add_artist( circle4 ) 
axes.add_artist( circle3 ) 
axes.add_artist( circle2 ) 
axes.add_artist( circle1 )

for num in throws1000:
  point = plt.Circle(throws1000.get(num), 0.01, color = "red" )
  axes.add_artist( point )



axes.set_xlim([-1.2,1.2])
axes.set_ylim([-1.2,1.2])

plt.title( 'Darts scoreboard' ) 
plt.show() 
```
