**[Problem 1] Creating random numbers**

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0) 

mean = (-3, 0)
cov = [[1.0, 0.8], [0.8, 1.0]]
x = np.random.multivariate_normal(mean, cov, (500))

print(x.shape)
print(x)
```

**[Problem 2] Visualization by scatter plot**

```python
import matplotlib.pyplot as plt

x1 = x[:,0]
x2 = x[:,1]
plt.scatter(x1,x2, label="scatter")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend(loc='upper left')
plt.show()
```

**[Problem 3] Visualization by histogram**

```python
# the histogram of the data
plt.xlabel('x1')
plt.ylabel('frequency')
plt.title('Histogram of x1')
plt.hist(x1)

plt.xlim(-5, 2)
#plt.ylim(0, 30)
#plt.grid(True)
plt.show()

plt.xlabel('x2')
plt.ylabel('frequency')
plt.title('Histogram of x2')
plt.hist(x1)

plt.xlim(-5, 2)
#plt.ylim(0, 30)
#plt.grid(True)
plt.show()
```

**[Problem 4] Addition of data**

```python
mean = (0, -3)
cov = [[1.0, 0.8], [0.8, 1.0]]
z = np.random.multivariate_normal(mean, cov, (500))

print(z.shape)
#print(z)

x1 = x[:,0]
x2 = x[:,1]
plt.scatter(x1,x2, label="0")
plt.xlabel("x1")
plt.ylabel("x2")

z1 = z[:,0]
z2 = z[:,1]

plt.scatter(z1,z2, label="1")
plt.legend(loc='upper left')

plt.show()
```

**[Problem 5] Data combination**

```python
merged = np.concatenate((x,z))
print(merged.shape)
```

**[Problem 6] Labeling**

```python
#creating labels
label1 = np.full((500, 1), 0)
#print(label1)
label2 = np.full((500, 1), 1)
#print(label1)

data1 = np.append(x, label1, axis=1)
print(data1.shape)
print(data1)

data2 = np.append(z, label2, axis=1)
print(data2.shape)
print(data2)

data = np.concatenate((data1,data2))
print(data.shape)
print(data)
```
